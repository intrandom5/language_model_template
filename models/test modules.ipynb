```python
import torch

inputs = torch.IntTensor([[87, 49, 32, 17, 64, 11, 18, 739],
                      [74, 27, 31, 11, 32, 0, 0, 0],
                      [2, 51, 38, 27, 0, 0, 0, 0]]) # 배치 내의 최대 문장 길이는 8. 3개의 문장.

positions = torch.IntTensor([i for i in range(inputs.shape[-1])]).repeat(3, 1)

print(inputs)
print(positions)
```

test transformers

```python
from Transformers import Transformers

transformer = Transformers(
    token_len=800,
    pad_id=0,
    n_encoder_layer=6,
    n_decoder_layer=6,
    h_dim=512,
    d_ff=2048
)

transformer
```

```python
outputs = transformer(inputs, inputs)
print(outputs.shape)
```

```python
from GPT import GPT

gpt = GPT(
    token_len=800,
    pad_id=0,
    n_layer=12,
    h_dim=768,
    num_heads=12,
    d_ff=3072,
    dropout=0.1,
    seq_max_len=512,
)

gpt
```

```python
outputs = gpt(inputs, positions)
print(outputs.shape)
```
